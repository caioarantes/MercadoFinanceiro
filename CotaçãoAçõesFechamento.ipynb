```python
import pandas as pd
from datetime import date
import yfinance as yf
from yahoofinancials import YahooFinancials
```

```python
df = pd.DataFrame(columns=['Empresa', 'Data', 'Cotação', 'Variação'])
df
```

```python
empresas = [
'SUZB3',
'KLBN11',
'RAIL3',
'SANB11',
'NTCO3',
'PETR4',
'ELET3',
'NUBR33',
'A1BM34',
'DEEC34',
'SBSP3',
'Z2EN34',
'UCAS3',
'A1ZN34',
'B3SA3',
'SLCE3',
'AGRO3',
'MDIA3',
'USIM5',
'CSNA3',
]

for codigo in empresas:
    try:
        df_aux = yf.download(codigo+'.SA', 
                              start='2022-05-01', 
                              #end='2022-08-26', até mais atual
                              progress=False,
        )
        df_aux = df_aux.reset_index(level=0)

        data = df_aux.iloc[-1:]['Date'].astype(str).tolist()[0]
        fechamento = df_aux['Close'].tolist()[-1]
        fechamento_anterior = df_aux['Close'].tolist()[-2]
        variacao = (fechamento/fechamento_anterior-1)
        df = df.append({'Empresa':codigo, 'Data':data, 'Cotação':fechamento, 'Variação':variacao}, ignore_index=True)
    except:
        df = df.append({'Empresa':codigo, 'Data':'erro', 'Cotação':'', 'Variação':''}, ignore_index=True)
```
